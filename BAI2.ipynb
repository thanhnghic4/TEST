```python
from mtcnn.mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = 'C:/Users/ADMIN/Pictures/anh_test.jpg'
# tao ham xem anh
def show_image(array):
    plt.imshow(cv2.cvtColor(array, cv2.COLOR_BGR2RGB))
```

```python
# doc anh
img = cv2.imread(path)
show_image(img)
```

```python
# nhan dang bang MTCNN
detector = MTCNN()
detect_bound = detector.detect_faces(img)
```

```python
# luu ket qua nha dang vao 1 list
list_face = []
for i in range(0,len(detect_bound)):
    list_face.append(detect_bound[i]['box'])
```

```python
list_face
```

```python
#hien thi cac anh

plt.axis('off')
k = 1
for face in list_face:
    plt.figure(k)
    k += 1
    crop = img[face[1]:face[1]+face[3],face[0]:face[0]+face[2],:]
    show_image(crop)
```
