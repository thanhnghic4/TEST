```python
import numpy as np
import cv2
import os
import random
from matplotlib import pyplot as plt
path = os.path.abspath(os.path.join('DATA'))
path = 'E:\Juyter Folder\BAI6\data'

def convert(path):
    image = cv2.imread(path)
    image.resize(48,48)
    return image
def Read_data(path):
    os.chdir(path)
    allfile = os.listdir('.')
    label = []
    train = []
    k = 0
    for file in allfile:
        path2 = os.path.join(path, file)
        os.chdir(path2)
        allname = os.listdir('.')
        for name in allname:
                path3 = os.path.join(path2, name)
                vt = convert(path3)
                train.append(vt)
                label.append(file)
    X_train = np.asarray(train)
    X_label = np.asarray(label)
    return X_train,X_label
```

```python
X_train,X_label = Read_data(path)
```

```python
def KNN(x,X_train,X_label):
    image = cv2.cvtColor(x,cv2.COLOR_BGR2GRAY)
    image.resize(48,48)
    d = np.abs(X_train - image)
    distance = []
    for i in d:
        distance.append(i.mean())
    mmin = np.argmin(distance)
    return X_label[mmin]
```
