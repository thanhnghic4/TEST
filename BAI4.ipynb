```python
from mtcnn.mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
path1 = 'C:/Users/ADMIN/Documents/FaceDB/IMG_5326.jpg'
path2 = 'C:/Users/ADMIN/Documents/FaceDB/IMG_5327.jpg'
path3 = 'C:/Users/ADMIN/Documents/FaceDB/IMG_5328.jpg'
path4 = 'C:/Users/ADMIN/Documents/FaceDB/IMG_5329.jpg'
path5 = 'C:/Users/ADMIN/Documents/FaceDB/IMG_5330.jpg'
path6 = 'C:/Users/ADMIN/Documents/FaceDB/IMG_5331.jpg'
path7 = 'C:/Users/ADMIN/Documents/FaceDB/IMG_5332.jpg'
# tao ham xem anh
def show_image(array):
    plt.imshow(cv2.cvtColor(array, cv2.COLOR_BGR2RGB))
path = np.array([path1,path2,path3,path4,path5,path6,path7])
```

```python
# doc anh
img  = []

for p in path:
    k = cv2.imread(p)
    print(k)
    img.append(k)    
```

```python
# nhan dang bang MTCNN
detector = MTCNN()
list_face = []
k = 1
for iimg in img:
    print(k)
    detect_bound = (detector.detect_faces(iimg))
    list_face = []
    for i in range(0,len(detect_bound)):
        list_face.append(detect_bound[i]['box'])
    print()
    for face in list_face:
        plt.figure(k)
        crop = iimg[face[1]:face[1]+face[3],face[0]:face[0]+face[2],:]
        show_image(crop)
        k += 1
```
